## Motivation

As we move forward, we will be interested in functions known as probability density functions (PDFs) and cumulative distributions functions (CDFs). Analyzing these completely requires calculus. Let's introduce these notions early, and try to understand them informally, so as we formalize our math, we can tell a story. Jumping ahead, if we have a probability density function $p(x)$, then the two relationships we will have will be

- $ C(x) = \int_a^x p(t)\ dt$, where $C(x)$ denotes the CDF of $p(x)$
- $\frac{d}{dx}\left[C(x)\right] = p(x)$

# A look at normal distributions

For the following, let's look at data of the heights of 25000 diffrent individuals. 

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 9)

df = pd.read_csv("hw_25000.csv")
df_height=df['Height']
print("Minimum heigth: ",df_height.min())
print("Maximum height: ",df_height.max(),'\n')
print("first few lines:",df_height.head(),'\n')
plt.xticks([x for x in range(int(df_height.min()),int(df_height.max())+1)])
s_height=sns.distplot(df_height, bins =20,kde=False,ax =ax)

```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 9)
s_height=sns.distplot(df_height, fit_kws={"color":"red"},bins =20,kde=True,hist=False,ax =ax)


```

```python
figb, ax = plt.subplots()
figb.set_size_inches(14, 9)
plt.xticks([x for x in range(int(df_height.min()),int(df_height.max())+1)])
s = pd.Series(df_height)
sns.kdeplot(s, cumulative=True)

```
